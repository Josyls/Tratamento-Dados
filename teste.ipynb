```python
sns.histplot(data=df, x='Idade', kde=True, bins=20)
plt.title('Distribuição frequencia Idade')
```

```python
sns.histplot(data=df, x='Renda Anual (em $)', kde=True, bins=4)
plt.title('Distribuição frequencia Renda Anual (em $)')
```

```python
tempoMax = df['Tempo no Site (min)'].max()
tempoMin = df['Tempo no Site (min)'].min()
usuariosObservados = df['Tempo no Site (min)'].count()
```

```python

print((tempoMax-tempoMin)/usuariosObservados)
```

```python
sns.histplot(data=df, x='Tempo no Site (min)', kde=True, bins=16)
plt.title('Distribuição frequencia Tempo no Site (min)')
```

```python
colunasComparadas = ['Idade', 'Renda Anual (em $)', 'Tempo no Site (min)', 'Compra (0 ou 1)']
sns.pairplot(df[colunasComparadas], hue='Compra (0 ou 1)', kind='kde')
plt.show
```

```python
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Gênero', hue='Compra (0 ou 1)')
```
